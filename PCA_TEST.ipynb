```python
from sklearn.preprocessing import StandardScaler
```

```python
scaler=StandardScaler()
scaler.fit(data)
```

```python
scaled_data=scaler.transform(data) 
```

```python
scaled_data
```

```python
scaled_data.shape
```

```python
from sklearn.decomposition import PCA
```

```python
pca=PCA(n_components=4)
```

```python
pca.fit(scaled_data)
```

```python
x_pca=pca.transform(scaled_data)    
```

```python
import matplotlib.pyplot as plt
import numpy as np


np.random.seed(0)
x_pca = np.random.randn(1000, 3) 

plt.figure(figsize=(8, 6))

plt.scatter(x_pca[:, 0], x_pca[:, 1], color='blue', label='First Principal Component')

plt.scatter(x_pca[:, 1], x_pca[:, 0], color='red', label='Second Principal Component')

plt.xlabel('First Principal Component')
plt.ylabel('Second Principal Component')
plt.title('Scatter Plot of Principal Components')
plt.legend()

plt.show()
```

```python
pca.explained_variance_ratio_
```

```python
import matplotlib.pyplot as plt
import numpy as np

variance = np.array([0.52331883, 0.33437727, 0.0761982, 0.03467531])

plt.figure(figsize=(8, 6))
plt.bar(range(1, len(variance) + 1), variance, alpha=0.7)
plt.xlabel('Principal Component')
plt.ylabel('Variance')
plt.title('Variance of Principal Components')
plt.xticks(range(1, len(variance) + 1))
plt.grid(axis='y')

plt.show()
```

```python
# second column to reconstruct..
column_to_reconstruct = 1  

# Retrieve the appropriate number of components for reconstruction
num_components_used = pca.n_components_  # Use the total components obtained from PCA
components_to_use = pca.components_[:num_components_used, :]

# Transpose the components matrix before the dot product
components_transpose = components_to_use.T

# Reconstructing the selected column using the principal components
reconstructed_column = np.dot(x_pca[:, :num_components_used], components_transpose)[:, column_to_reconstruct]

# Rescale the reconstructed column if the data was scaled before PCA
reconstructed_column_rescaled = reconstructed_column * scaler.scale_[column_to_reconstruct] + scaler.mean_[column_to_reconstruct]

# Assuming 'data' is the original DataFrame before scaling
data_reconstructed = data.copy()  # Create a copy of the original data
data_reconstructed.iloc[:, column_to_reconstruct] = reconstructed_column_rescaled  # Replace the reconstructed column

# Save the reconstructed DataFrame to a CSV file
data_reconstructed.to_csv('reconstructed_data.csv', index=False)
```
